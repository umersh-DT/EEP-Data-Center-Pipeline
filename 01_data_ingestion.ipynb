## 📊 Executive Storyline: The Structural Turnaround of EEP

This chart visualizes the dramatic financial pivot of **Ethiopian Electric Power (EEP)** as high-density tech infrastructure and data centers integrated into the national power grid. By acting as reliable, continuous anchor customers, these data facilities fundamentally transformed the utility’s economic model.

### 🔍 Key Insights from the Data

* **The Loss-Making Era (Pre-2024):** 
  Historically, EEP faced staggering structural deficits, recording comprehensive net losses of **-26.2 Billion Birr** in FY 2021/22 and **-24.8 Billion Birr** in FY 2022/23. The utility suffered from under-monetized power generation assets because domestic distribution infrastructure lagged behind generation capabilities.
  
* **The Inflection Point (FY 2024/25):** 
  As private technology data centers expanded their operations across the grid, tech energy consumption surged exponentially to **7,960 GWh**—capturing a massive **27% share of EEP's total power output**. 

* **The Profitable Marriage:** 
  Because these data centers operate at a near-flat, maximum continuous load factor, they absorb massive baseload energy (such as the surplus generated by the Grand Ethiopian Renaissance Dam - GERD) that would otherwise go to waste. More importantly, policy shifts allowed EEP to bill these institutional tech clients directly in foreign exchange, injecting **$220 Million USD** in hard currency into EEP’s revenues.

### 💡 The Macroeconomic Conclusion

This transformation drove an unprecedented financial swing, lifting EEP into positive territory with a **5.2 Billion Birr net profit** in FY 2024/25 and a projected **14.2 Billion Birr profit** for FY 2025/26. 

> **The Portfolio Narrative:** For government stakeholders and public utilities, this data proves that welcoming tech infrastructure is not an energy drain. Instead, it is the ultimate financial engine. The immense revenue and hard foreign currency generated by these facilities provide the state with the exact capital needed to fund grid modernization, transmission repairs, and rural electrification across Ethiopia.


```python
import os
from pathlib import Path
import pandas as pd
import numpy as np

# 1. Standard Path Setup
NOTEBOOK_DIR = Path(os.getcwd())
PROJECT_ROOT = NOTEBOOK_DIR.parent if NOTEBOOK_DIR.name == "notebooks" else NOTEBOOK_DIR
RAW_DATA_PATH = PROJECT_ROOT / "data" / "raw" / "eep_baseline.csv"
PROCESSED_DATA_DIR = PROJECT_ROOT / "data" / "processed"

# Ensure processed directory exists
PROCESSED_DATA_DIR.mkdir(parents=True, exist_ok=True)

# 2. Ingest original baseline
if RAW_DATA_PATH.exists():
    df_eep = pd.read_csv(RAW_DATA_PATH)
else:
    raise FileNotFoundError(f"Baseline file missing at {RAW_DATA_PATH}")

# 3. Simulate Customs Data (ECC Layer)
# Hardware imports typically scale up slightly before massive grid draws happen
customs_data = {
    "Fiscal_Year": ["2021/22", "2022/23", "2023/24", "2024/25", "2025/26 (Proj)"],
    "Est_Hardware_Import_Value_Million_USD": [0.0, 8.5, 55.0, 195.0, 260.0],
    "Avg_Customs_Duty_Tariff_Pct": [5.0, 5.0, 5.0, 5.0, 5.0],  # Modeling a steady capital goods concession tariff rate
}
df_ecc = pd.DataFrame(customs_data)

# 4. Feature Engineering: Calculate Upfront Gov Revenue & Lead Indicators
# Upfront revenue collected at customs clearance point
df_ecc["Upfront_Customs_Revenue_Million_USD"] = (
    df_ecc["Est_Hardware_Import_Value_Million_USD"] * (df_ecc["Avg_Customs_Duty_Tariff_Pct"] / 100.0)
)

# 5. Merge the Datasets (EEP Utility + ECC Customs)
df_merged = pd.merge(df_eep, df_ecc, on="Fiscal_Year")

# 6. Model the 90-day (1-Quarter) Infrastructure Lag
# Shift hardware imports down by 1 period to represent the timeline between border arrival and grid power connection
df_merged["Leading_Hardware_Import_Value_Next_Year_FDI"] = df_merged["Est_Hardware_Import_Value_Million_USD"].shift(-1)

# 7. Save the fully enriched dataset to the processed layer
output_file = PROCESSED_DATA_DIR / "eep_ecc_integrated_metrics.csv"
df_merged.to_csv(output_file, index=False)

print(f"🎉 Success! Multi-Agency Integrated Pipeline Saved to: {output_file}")
display(df_merged[["Fiscal_Year", "Net_Income_Loss_Billion_Birr", "Est_Hardware_Import_Value_Million_USD", "Upfront_Customs_Revenue_Million_USD"]])
```
